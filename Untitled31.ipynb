# Data Cleaning And Stats

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data.shape
```

```python
data['Rate'].value_counts()
```

```python
data['len']=data['Comment'].str.len()
data
```

```python
data['Comment']=data['Comment'].str.lower()
data
```

```python
data['Comment']=data['Comment'].str.replace(r'â‚¹', 'rupees')
data['Comment']=data['Comment'].str.replace(r'\d+(\.\d+)?', 'numbers')
```

```python
data['Comment']=data['Comment'].str.replace(r'[^\w\d\s]', ' ')
data['Comment']=data['Comment'].str.replace(r'\s+', ' ')
data['Comment']=data['Comment'].str.replace(r'^\s+|\s+?$','')
data
```

```python
import string
import nltk
from nltk.corpus import stopwords
stopwords=set(stopwords.words('english')+['u', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure'])
data['Comment']=data['Comment'].apply(lambda x: ' '.join(term for term in x.split() if term not in stopwords))
```

```python
data['clean_length']=data.Comment.str.len()
data.head()
```
